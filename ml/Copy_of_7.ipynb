   Train clustring with the K-Means algorithm using suitable dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Iris.csv")

# X contains input features
X = df.drop("Species", axis=1)

# Feature scaling (important for K-means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# Cluster labels
labels = kmeans.labels_
print("Cluster Labels:", labels)

# Cluster centers
centroids = kmeans.cluster_centers_
print("Cluster Centers:\n", centroids)

# Plot using first two features (for visualization)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200)
plt.xlabel("Sepal Length (scaled)")
plt.ylabel("Sepal Width (scaled)")
plt.title("K-Means Clustering on Iris Dataset")
plt.show()
```

```python
# Plot clustering result using feature 2 and 3
plt.scatter(X_scaled[:, 2], X_scaled[:, 3], c=labels)
plt.scatter(centroids[:, 2], centroids[:, 3], marker='x', s=200)

plt.xlabel("Petal Length (scaled)")
plt.ylabel("Petal Width (scaled)")
plt.title("K-Means Clustering (Petal Features)")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

print("Inertia (WCSS):", kmeans.inertia_)
print("Silhouette Score:", silhouette_score(X_scaled, labels))
```
