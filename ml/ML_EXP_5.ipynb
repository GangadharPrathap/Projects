```python
from sklearn.model_selection import train_test_split

x=df.drop('target',axis=1)
y=df['target']

x_train, x_test, y_train, y_tst = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
log_reg_model=LogisticRegression()
log_reg_model.fit(x_train,y_train)
print('LogisticRegression successfully completed')
```

```python
x_test.shape
```

```python
y_pred = log_reg_model.predict(x_test)
print('prediction on test data is successfully compeleted')
```

```python
predictions_df = pd.DataFrame({'Actual':y_test,"predicted":y_pred})
predictions_df['Correct']=(predictions_df['Actual']==predictions_df['predicted'])
display(predictions_df.head())
```

```python
from sklearn.metrics import(
    confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
)
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)

accuracy=accuracy_score(y_test,y_pred)
precision=precision_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
f1_score=f1_score(y_test,y_pred)

print("\n==performance of logistic regression model==")
print(f"accuracy={accuracy*100:.2f}")
print(f"precision={precision*100:.2f}")
print(f"recall={recall*100:.2f}")
print(f"f1_score={f1_score*100:.2f}")
```
