   Train clustring with the K-Means algorithm using suitable dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("/content/heart (1).csv")
```

```python
df.head()
```

```python
X= df.drop('target',axis = 1 )
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

kmeans = KMeans(n_clusters = 3 ,random_state = 42)
kmeans.fit(X_scaled)
```

```python
labels = kmeans.labels_
print("Cluster labels:",labels)
```

```python
Centroids = kmeans.cluster_centers_
print("Cluster Centers:\n",Centroids)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,3],c=labels)
plt.scatter(Centroids[:,0],Centroids[:,3],marker = 'x',s=200)
plt.xlabel("Age (Scaled)")
plt.ylabel("Trestbps (Scaled)")
plt.title("K-Means Clustering on Heart Dataset")
plt.show()
```

```python
plt.scatter(X_scaled[:,4],X_scaled[:,7],c=labels)
plt.scatter(Centroids[:,4],Centroids[:,7],marker = 'x',s=200)
plt.xlabel("chol (Scaled)")
plt.ylabel("thalach (Scaled)")
plt.title("K-Means Clustering on Heart Dataset")
plt.show()
```
